```python
def convert(row):
        try:
            time = str(row["time"]) if str(row["time"]) != "2400" else "0000"
            x = str(row["date"]) + " " + time
            return datetime.strptime(x, '%Y%m%d %H%M')
        except:
            print(x)
```

```python
frame["datetime"] = frame.apply(lambda x: convert(x), axis=1)
```
